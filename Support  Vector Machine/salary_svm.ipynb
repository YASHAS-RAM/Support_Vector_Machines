```python
from sklearn.preprocessing import LabelEncoder
```

```python
train = train.apply(LabelEncoder().fit_transform)
```

```python
train.head()
```

```python
test = test.apply(LabelEncoder().fit_transform)
```

```python
train.hist(edgecolor='red',figsize=(15,12),linewidth=1) 
plt.show()
```

```python
test.hist(edgecolor='red',figsize=(15,12),linewidth=1) 
plt.show()
```

```python
drop_elements = ['education', 'native', 'Salary']
X = train.drop(drop_elements, axis=1)
y = train['Salary']
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
corr=train.corr()
```

```python
from typing import ChainMap
fig, ax = plt.subplots(figsize=(15, 8))
sns.heatmap(corr, cmap='magma', annot=True, fmt=".5f")
plt.xticks(range(len(corr.columns)), corr.columns)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.show()
```

```python
from sklearn import metrics
from sklearn import svm
from sklearn.svm import SVC
svc=SVC()
```

```python
svc.fit(X_train, y_train)
```

```python
pred = svc.predict(X_test)
```

```python
metrics.classification_report(y_test, pred)
```

```python
metrics.confusion_matrix(y_test, pred)
```

```python
metrics.accuracy_score(y_test,pred)
```

```python
#This model is 80% accurate
```

```python
metrics.precision_score(y_test,pred)
```

```python
#model is good at predicting specific category with 86% as precision score
```

```python
metrics.recall_score(y_test,pred)
```

```python
#28% of time model was abel to detect specific category
```
